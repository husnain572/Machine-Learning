# Feature encoding

```python
# import libraries
import pandas as pd

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,OrdinalEncoder
```

```python
le=LabelEncoder()
df['encoded_time']=le.fit_transform(df['time'])
```

```python
df.tail(100)
```

```python
df['encoded_time'].value_counts()
```

### Using Ordinal encoder for encoding `day` column

```python
df['day'].value_counts()
```

```python
Oe=OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
df['encoded_day']=Oe.fit_transform(df[['day']])
df.head()
```

```python
df['encoded_day'].value_counts()
```

# One hot encoding on `sex` column

```python
oh=OneHotEncoder()
oh.fit_transform(df[['sex']]).toarray()
```
